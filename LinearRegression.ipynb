```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


plt.rcParams['figure.figsize'] = [20.0, 10.0]


url = 'https://raw.githubusercontent.com/mubaris/potential-enigma/master/headbrain.csv'
data = pd.read_csv(url)
print(data.shape)
data.head()
```

```
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values

mean_x = np.mean(X.copy())
mean_y = np.mean(Y.copy())

n = len(X)

num = 0
den = 0

for i in range(n):
  num += (X[i] - mean_x) * (Y[i] - mean_y)
  den += (X[i] - mean_x) ** 2

b1 = float(num) / float(den)
b0 = mean_y - (b1 * mean_x)

print(b0, b1)
```

```
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, label='Linier Regression')
plt.scatter(X, Y, label='Scatter Plot')

plt.xlabel('Head Size')
plt.ylabel('Brain Weight')

plt.legend()
plt.show()
```

```
ss_t = 0
ss_r = 0

for i in range(n):
  y_pred = b0 + b1 * X[i]
  ss_t += (Y[i] - mean_y) ** 2
  ss_r += (Y[i] - y_pred) ** 2

r2 = 1 - (ss_r/ss_t)
print(r2)
```
